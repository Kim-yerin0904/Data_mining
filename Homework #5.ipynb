```python
import itertools
features = [ 'SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth' ]

combinations = itertools.combinations(features, 2)
fig = plt.figure(figsize=(15,10))
figure_index = 1
groups = iris.groupby('Name')

for combination in combinations:
    axs = fig.add_subplot(2,3,figure_index)
    for name,group in groups:
        axs.plot(group[combination[0]],group[combination[1]],marker='o',linestyle='',label=name)
    axs.set_xlabel(combination[0])
    axs.set_ylabel(combination[1])
    figure_index+=1
plt.legend()
plt.show()
```

### 3단계

```python
plt.figure(figsize=(10,10))

for name,group in groups:
    plt.plot(group["PetalLength"],group["PetalWidth"],marker='o',linestyle='',label=name)
# plot the decision boundary for # iris-setosa
plt.fill([0.5,0.5,7.0, 7.0],[0.0, 0.7, 0.7, 0.0],color='blue',alpha=0.1)
plt.fill([3.0,3.0,4.8, 4.8],[0.9, 1.6, 1.6, 0.9],color='orange',alpha=0.1)
plt.fill([4.8,4.8,7.0, 7.0],[1.5, 2.5, 2.5, 1.5],color='green',alpha=0.1)

#finer grid (optional)
plt.minorticks_on()
plt.grid(which='both')
# set axis labels
plt.xlabel("Petal length")
plt.ylabel("Petal width")
# add legend
plt.legend()
# show the plot
plt.show()
```

### 4단계

- setosa: 0.5 <= PetalLength <= 7 AND 0 <= PetalWidth <= 0.7
- versicolor: 3.0 <= PetalLength <= 4.8 AND 0.9 <= PetalWidth <= 1.6
- virginica: 4.8 <= PetalLength <= 7 AND 1.5 <= PetalWidth <= 2.5
